```python

import pandas as pd
df1=pd.read_csv("SOCR-HeightWeight.csv")
print(df1)
```

```python
import matplotlib.pyplot as plt
import plotly.express as px
```

```python
height=df1['Height(Inches)']
weight=df1['Weight(Pounds)']
print("Mean Height of the class: ",height.mean())
print("Mean Weight of the class: ",weight.mean())
print("Median Height of the class: ",height.median())
print("Median Weight of the class: ",weight.median())
print("Mode Height of the class: ",height.mode())
print("Mode Weight of the class: ",weight.mode())
plt.scatter(height,weight,color='r')
plt.show()
#plt.bar(height,weight)
figure=px.scatter(height,weight)
figure.show()
```

This can be used to have a statistical survey of height and weigt for medical and other related researches like Census.
